```python
import cv2 as cv
import matplotlib.pyplot as plt
import numpy as np
import os
```

# Dilation

```python
img = np.random.rand(40, 40)
binary_img = (img > 0.95).astype(np.uint8) * 255
plt.figure(figsize=(10,5))
plt.subplot(121)
plt.imshow(img, cmap="gray", vmin=0, vmax=1)
plt.subplot(122)
plt.imshow(binary_img, cmap="gray", vmin=0, vmax=255)
```

```python
kernel = np.array([[1,1,1], [1,1,1], [1,1,1]]) # np.full, np.ones
img_dilate = cv.dilate(binary_img, kernel, iterations=10)
```

```python
plt.figure(figsize=(10,5))
plt.subplot(121)
plt.imshow(binary_img, cmap="gray", vmin=0, vmax=255)
plt.subplot(122)
plt.imshow(img_dilate, cmap="gray", vmin=0, vmax=255)
```

```python
img = np.random.rand(40, 40)
binary_img = (img > 0.05).astype(np.uint8) * 255
img_erode = cv.erode(binary_img, kernel, iterations=1)
```

```python
plt.figure(figsize=(10,5))
plt.subplot(121)
plt.imshow(binary_img, cmap="gray", vmin=0, vmax=255)
plt.subplot(122)
plt.imshow(img_erode, cmap="gray", vmin=0, vmax=255)
```

```python
img_path = os.path.join("data", "coins.jpg")
img = cv.imread(img_path)
img_color = cv.cvtColor(img, cv.COLOR_BGR2RGB)
img_gray = cv.cvtColor(img, cv.COLOR_BGR2GRAY)
```

```python
plt.imshow(img_gray, cmap="gray")
```

```python
img = np.copy(img_gray)
rows, cols = np.shape(img)
for _ in range(100):
    i = np.random.randint(0, rows)
    j = np.random.randint(0, cols)
    img[i,j] = 0
```

```python
plt.imshow(img, cmap="gray")
```

```python
t, binary_img = cv.threshold(img, 0, 255, cv.THRESH_OTSU + cv.THRESH_BINARY_INV)
```

```python
img_open = cv.morphologyEx(binary_img, cv.MORPH_OPEN, kernel, iterations=1)
```

```python
plt.figure(figsize=(10,5))
plt.subplot(121)
plt.imshow(binary_img, cmap="gray", vmin=0, vmax=255)
plt.subplot(122)
plt.imshow(img_open, cmap="gray", vmin=0, vmax=255)
```

```python
t, binary_img = cv.threshold(img, 105, 255, cv.THRESH_BINARY_INV)
plt.imshow(binary_img, cmap="gray")
```

```python
img_close = cv.morphologyEx(binary_img, cv.MORPH_CLOSE, kernel, iterations=5)
```

```python
plt.figure(figsize=(10,5))
plt.subplot(121)
plt.imshow(binary_img, cmap="gray", vmin=0, vmax=255)
plt.subplot(122)
plt.imshow(img_close, cmap="gray", vmin=0, vmax=255)
```

```python
t, binary_img = cv.threshold(img, 155, 255, cv.THRESH_BINARY_INV)
img_morph_grad = cv.morphologyEx(binary_img, cv.MORPH_GRADIENT, kernel, iterations=1)
plt.imshow(img_morph_grad, cmap="gray")
```
